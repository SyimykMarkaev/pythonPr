Moving Sum


```python
arr = [1, 2, 3, 7, 9]
moving_sum=[]
window_size=3
i=0
while i<len(arr)-window_size+1:
    moving_sum.append(np.round(np.sum(arr[i:i+window_size])/window_size,0))
    i+=1
```

Least Squares Regression 

```python
from scipy import optimize
plt.style.use('seaborn-poster')
%matplotlib inline
```

Direct inverse method


```python

x=np.linspace(0,1,101)
y=1+x+x*np.random.random(len(x))
A = np.vstack([x, np.ones(len(x))]).T
A = np.vstack([x, np.ones(len(x))]).T
alpha = np.dot((np.dot(np.linalg.inv(np.dot(A.T,A)),A.T)),y)
print(alpha)
plt.figure(figsize = (10,8))
plt.plot(x, y, 'b.')
plt.plot(x, alpha[0]*x + alpha[1], 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Pseudoinverse

```python
pinv = np.linalg.pinv(A)
alpha = pinv.dot(y)
print(alpha)

plt.figure(figsize = (10,8))
plt.plot(x, y, 'b.')
plt.plot(x, alpha[0]*x + alpha[1], 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
alpha = np.linalg.lstsq(A, y, rcond=None)[0]
print(alpha)

# plot the results
plt.figure(figsize = (10,8))
plt.plot(x, y, 'b.')
plt.plot(x, alpha[0]*x + alpha[1], 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = np.linspace(0, 1, 101)
y = 1 + x + x * np.random.random(len(x))

def func(x, a, b):
    y = a*x+b
    return y

alpha = optimize.curve_fit(func, xdata = x, ydata = y)[0]
print(alpha)

# plot the results
plt.figure(figsize = (10,8))
plt.plot(x, y, 'b.')
plt.plot(x, alpha[0]*x + alpha[1], 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Group all years in the 10 years periods 

```python
start=1985
end=2016
def decade_year(col):
    if col<=1990:
        return "1981-1990"
    if col<=2000:
        return "1991-2000"
    if col <= 2010:
        return "2001=2010"
    else:
        return "2011-2016"
    
```

```python
df['year']=df['year'].apply(decade_year)
```

```python
plt.figure(figsize=(12,7))
sns.countplot(df.year)
```

Population-> Stratify into groups

```python
def stratify_population(col):
    if col<=10000:
        return "0-10000"
    if col<=100000:
        return "10001-100000"
    if col<=1000000:
        return "100001-1000000"
    if col<=10000000:
        return "1000001-10000000"
    else:
        return ">10000000"
```

```python
df['population']=df['population'].apply(stratify_population)
```

```python
plt.figure(figsize=(12,7))
sns.countplot(x='population',data=df)
```

```python
def gdp_for_year(col):
    if len(col)<10:
        return "0-999.999.999"
    if len(col)<11:
        return "1.000.000.000  9.999.999.999"
    if len(col)<12:
        return "10.000.000.000  99.999.999.999"
    else:
        return ">100.000.000.000"
```

```python
def gdp_per_capita(col):
    if col<=100:
        return "0-100"
    if col<=1000:
        return "101  1.000"
    if col<=10000:
        return "1.001  10.000"
    if col<=50000:
        return "10.001  50.000"
    if col<=100000:
        return "50.001  100.000"
    else:
        return "> 100.000"
```

```python
df['gdp_per_capita ($)']=df['gdp_per_capita ($)'].apply(gdp_per_capita)
```
